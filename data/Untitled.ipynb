```python
## GoodReads Scraping

import requests
import re
import pandas as pd
import bs4
import lxml
url = 'https://www.goodreads.com/book/most_read?category=all&country=all&duration=m'
resp = requests.get(url)
resp.content[:100]

soup = bs4.BeautifulSoup(resp.text, "html.parser")
soup.find("a", class_="bookTitle").text.strip()
table = soup.find_all('table')[0]
table

```

```python
#titles
titles = [x.text.strip() for x in soup.find_all("a", class_="bookTitle")]
len(titles)
```

```python
# reads
readers_el = soup.find_all("span", "statistic")
readers = [''.join(re.findall(r"(\d+,\d+)", x.text)) for x in readers_el]
readers
```

```python
#author names
rows = table.find_all('a', "authorName")
rows = [row.text.strip() for row in rows]
len(rows) 

#Remove 'Want to read', published, and spaces
#Remove out of stars
```

```python
#rating out of 5 stars
rating = ["".join(re.findall(r"(\d\.\d{1,2})",x.text)) for x in soup.find_all("span", "minirating")]
len(rating)
```

```python
ratn_reviews = ["".join(re.findall(r"(\d+,\d+ | \d+,\d+,\d+)",x.text)).strip() for x in soup.find_all("span", "minirating")]
len(ratn_reviews)
```

```python
data = pd.DataFrame ({"title": titles, "author": rows, "read": readers, "reviewed": ratn_reviews, "stars": rating})
data.head()
```

```python
data.to_csv("../data/GR.csv", index=False)
```
